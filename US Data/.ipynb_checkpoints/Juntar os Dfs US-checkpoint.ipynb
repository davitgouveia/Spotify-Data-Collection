```python
import pandas as pd
```

## Juntando os Dataframes

```python
base = "spotify23-US-"
bancofinal = pd.DataFrame()
for i in range(1,16):
    banco = pd.read_csv(base+str(i)+".csv")
    bancofinal = bancofinal.append(banco)
display(bancofinal)
```

## Dropando os NaN

```python
bancofinal = bancofinal.dropna()
```

## Dropando os Duplicados

```python
bancofinal = bancofinal.drop_duplicates()
bancofinal
```

## Criando uma coluna market

```python
bancofinal["market"] = "US"
bancofinal
```
