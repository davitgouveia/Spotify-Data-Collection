# Dataframe treatment

## Import the libraries

```python
import pandas as pd
```

## Joining the Dataframes

```python
base = "spotify23-US-"
df_final = pd.DataFrame()
for i in range(1,16):
    df = pd.read_csv(base+str(i)+".csv")
    df_final = df_final.append(df)
display(df_final)
```

## Dropping the NaN

```python
df_final = df_final.dropna()
```

## Dropping Duplicates

```python
df_final = df_final.drop_duplicates()
df_final
```

## Creating a market column

```python
df_final["market"] = "US"
df_final
```

## Remove column Unnamed: 0

```python
df_final = df_final.drop("Unnamed: 0", axis = 1)
df_final
```

## Creates the df_final file

```python
df_final.to_csv("df_final_US.csv")
```
